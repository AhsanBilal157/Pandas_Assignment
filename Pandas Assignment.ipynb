```python
import pandas as pd
```

```python
 import os
from glob import glob

files=glob('states*')
uc_census = pd.concat((pd.read_csv(file) for file in files),ignore_index = True)
uc_census
    
```

```python
uc_census.dtypes
```

```python
uc_census['Income'] = uc_census['Income'].replace("\$","",regex=True)
uc_census['Income'] = uc_census['Income'].astype('float64')
uc_census['Income']
```

```python
uc_census[["Male","Female"]] = uc_census['GenderPop'].str.split("_",expand = True)
del uc_census['GenderPop']
```

```python
uc_census
```

```python
uc_census["Male"]= uc_census["Male"].str.extract("(\d+)", expand=True)
uc_census["Female"]=uc_census["Female"].str.extract("(\d+)", expand=True)
uc_census["Male"]=pd.to_numeric(uc_census["Male"])
uc_census["Female"]=pd.to_numeric(uc_census["Female"])
```

```python
import matplotlib.pyplot as plt
plt.scatter(uc_census['Female'],uc_census['Income'])
plt.show()
```

```python
value = uc_census["TotalPop"]-uc_census["Male"]
uc_census['Female']=uc_census['Female'].fillna(value= value)
```
